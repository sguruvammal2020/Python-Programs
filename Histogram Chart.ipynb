```python
# To differentiate from bar plot use histtype
import matplotlib.pyplot as plt
no=[47,13,52,16,15,58,57,32,35,13,56,46,42,47,37,23,33,50,32,14,23,35,52,16,51,
    47,22,22,16,24,54,13,16,45,39,58,38,19,58,25,25,25,40,44,10,42,50,23,35,27]
l=[10,20,30,40,50,60]          # Bins parameter
plt.hist(no,color="r",bins=l,edgecolor="g",histtype="step")   # or barstacked/bar/stepfilled
plt.title("Histogram Chart")
plt.xlabel("Python")
plt.ylabel("No.")
plt.show()
```
