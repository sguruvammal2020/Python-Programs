```python
import matplotlib.pyplot as plt
x=["python","C","C++","Java"]     # list of x values
y=[85,70,60,82]    # list of y values
plt.bar(x,y)
plt.title("Programming Language Vs Popularity",fontsize=15)   # Title for the plot
plt.xlabel("Language",fontsize=10)                             # Name of X axis
plt.ylabel("Popularity",fontsize=10)                           # Name of Y axis
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["python","C","C++","Java"]     # list of x values
y=[85,70,60,82]    # list of y values
plt.bar(x,y,width=0.2)                                       # width for bar
plt.title("Programming Language Vs Popularity",fontsize=15)   # Title for the plot
plt.xlabel("Language",fontsize=10)                             # Name of X axis
plt.ylabel("Popularity",fontsize=10)                           # Name of Y axis
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["python","C","C++","Java"]     # list of x values
y=[85,70,60,82]    # list of y values
plt.bar(x,y,width=0.2,color="r")                       # Change the default color from Blue to Red
plt.title("Programming Language Vs Popularity",fontsize=15)   # Title for the plot
plt.xlabel("Language",fontsize=10)                             # Name of X axis
plt.ylabel("Popularity",fontsize=10)                           # Name of Y axis
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["python","C","C++","Java"]     # list of x values
y=[85,70,60,82]    # list of y values
c=["y","m","g","r"]
plt.bar(x,y,width=0.4,color=c)                       # set different color to different bar
plt.title("Programming Language Vs Popularity",fontsize=15)   # Title for the plot
plt.xlabel("Language",fontsize=10)                             # Name of X axis
plt.ylabel("Popularity",fontsize=10)                           # Name of Y axis
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["python","C","C++","Java"]     # list of x values
y=[85,70,60,82]    # list of y values
c=["y","m","g","b"]
plt.bar(x,y,width=0.4,color=c,align="edge")         # Set the bar to the edge of tick of X axis (but default: align="centre")
#plt.bar(x,y,width=0.4,color=c,align="center")
plt.title("Programming Language Vs Popularity",fontsize=15)   # Title for the plot
plt.xlabel("Language",fontsize=10)                             # Name of X axis
plt.ylabel("Popularity",fontsize=10)                           # Name of Y axis
plt.show()
```

```python
# Add edgecolor to the bar
import matplotlib.pyplot as plt
x=["python","C","C++","Java"]     # list of x values
y=[85,70,60,82]    # list of y values
c=["y","m","g","r"]
plt.bar(x,y,width=0.4,color=c,align="edge",edgecolor="r",linewidth=5,linestyle=":",alpha=0.5)  
#plt.bar(x,y,width=0.4,color=c,align="center")
plt.title("Programming Language Vs Popularity",fontsize=15)   
plt.xlabel("Language",fontsize=10)                             
plt.ylabel("Popularity",fontsize=10)                           
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["python","C","C++","Java"]     # list of x values
y=[85,70,60,82]    # list of y values
c=["y","m","g","r"]
plt.bar(x,y,width=0.4,color=c,align="edge",edgecolor="r",label="Popularity 1")         
#plt.bar(x,y,width=0.4,color=c,align="center")
plt.legend()                                                            # Set the label of bar() as legend
plt.title("Programming Language Vs Popularity",fontsize=15)   # Title for the plot
plt.xlabel("Language",fontsize=10)                             # Name of X axis
plt.ylabel("Popularity",fontsize=10)                           # Name of Y axis
plt.show()
```

```python
# Double Bar plot
import matplotlib.pyplot as plt
x=["python","C","C++","Java"]     # list of x values
y=[85,70,60,82]    # list of y values
z=[20,30,40,50]   # Out of x=85, first 20 occupied by z=20
plt.bar(x,y,width=0.4,color="r",label="Popularity 1")         # Draw Bar 1 Full
plt.bar(x,z,width=0.4,color="y",label="Popularity 2")         # Draw Bar 2
plt.legend()                                                            
plt.title("Programming Language Vs Popularity",fontsize=15)   
plt.xlabel("Language",fontsize=10)                             
plt.ylabel("Popularity",fontsize=10)                           
plt.show()
```

```python
# Double Bar plot
import matplotlib.pyplot as plt
import numpy as np

x=["python","C","C++","Java"]     # list of x values
y=[85,70,60,82]    # list of y values
z=[20,30,40,50]   
p=np.arange(len(x))         # Arrange X axis values           
plt.bar(p,y,width=0.4,color="r",label="Popularity 1")         # Draw Bar 1 Full
plt.bar(p,z,width=0.4,color="y",label="Popularity 2")         # Draw Bar 2
plt.legend()                                                            
plt.title("Programming Language Vs Popularity",fontsize=15)   
plt.xlabel("Language",fontsize=10)                             
plt.ylabel("Popularity",fontsize=10)                           
plt.show()
```

```python
# Double Bar plot
import matplotlib.pyplot as plt
import numpy as np

x=["python","C","C++","Java"]     # list of x values
y=[85,70,60,82]    # list of y values
z=[20,30,40,50]  
width=0.2         # common width
p=np.arange(len(x))         # Arrange X axis values           
plt.bar(p,y,width,color="r",label="Popularity 1")         # Draw Bar 1 Full
plt.bar(p,z,width,color="y",label="Popularity 2")         # Draw Bar 2
plt.legend()                                                            
plt.title("Programming Language Vs Popularity",fontsize=15)   
plt.xlabel("Language",fontsize=10)                             
plt.ylabel("Popularity",fontsize=10)                           
plt.show()
```

```python
# Double Bar plot
import matplotlib.pyplot as plt
import numpy as np

x=["python","C","C++","Java"]     # list of x values
y=[85,70,60,82]    # list of y values
z=[20,30,40,50]  
width=0.2         # common width
p=np.arange(len(x))         # Arrange X axis values 
p1=[j+width for j in p]
plt.bar(p,y,width,color="r",label="Popularity 1")         # Draw Bar 1 Full
plt.bar(p1,z,width,color="y",label="Popularity 2")         # Draw Bar 2
plt.legend()                                                            
plt.title("Programming Language Vs Popularity",fontsize=15)   
plt.xlabel("Language",fontsize=10)                             
plt.ylabel("Popularity",fontsize=10)                           
plt.show()
```

```python
# Double Bar plot
import matplotlib.pyplot as plt
import numpy as np

x=["python","C","C++","Java"]     # list of x values
y=[85,70,60,82]    # list of y values
z=[20,30,40,50]  
width=0.2         # common width
p=np.arange(len(x))         # Arrange X axis values 
p1=[j+width for j in p]
plt.bar(p,y,width,color="r",label="Popularity 1")         # Draw Bar 1 Full
plt.bar(p1,z,width,color="y",label="Popularity 2")         # Draw Bar 2
plt.xticks(p+width/2,x,rotation=20)      # Rotate x axis values in 20 degree
plt.legend()                                                            
plt.title("Programming Language Vs Popularity",fontsize=15)   
plt.xlabel("Language",fontsize=10)                             
plt.ylabel("Popularity",fontsize=10)                           
plt.show()
```

```python
#Horizontal Bar Plot
import matplotlib.pyplot as plt
x=["python","C","C++","Java"]     # list of x values
y=[85,70,60,82]    # list of y values
z=[20,30,40,50] 
plt.title("Language Popularity Graph",fontsize=15)
plt.xlabel("Language",fontsize=10)
plt.ylabel("No.",fontsize=10)
plt.barh(x,y,color="r",label="popularity 1")
plt.barh(x,z,color="y",label="popularity 2")
plt.legend()
plt.show()
```
