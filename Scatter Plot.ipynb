<h1> Scatter Plot</h1>

```python
# Draw a scatter plot 
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
plt.scatter(day,no)
plt.show()
```

```python
# Draw a scatter plot 
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
plt.scatter(day,no,color="r")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=10)
plt.ylabel("No",fontsize=10)
plt.show()
```

```python
# Draw a scatter plot 
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]
sizes=[10,2,4,5,2,10,7]
plt.scatter(day,no,c=colors,s=sizes)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=10)
plt.ylabel("No",fontsize=10)
plt.show()
```

```python
# Draw a scatter plot 
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]
sizes=[400,200,300,100,200,100,300]
plt.scatter(day,no,c=colors,s=sizes,marker="*",alpha=0.4)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=10)
plt.ylabel("No",fontsize=10)
plt.show()
```

```python
# Draw a scatter plot 
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]
sizes=[400,200,300,100,200,100,300]
plt.scatter(day,no,c=colors,s=sizes,marker="*",edgecolor="g",linewidth=2)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=10)
plt.ylabel("No",fontsize=10)
plt.show()
```

```python
# Draw a scatter plot 
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=[10,49,30,29,56,20,30]
sizes=[400,200,300,100,200,100,300]
plt.scatter(day,no,c=colors,s=sizes,cmap="viridis")    # Color Map
plt.colorbar()
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=10)
plt.ylabel("No",fontsize=10)
plt.show()
```

```python
# Draw a scatter plot 
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=[10,49,30,29,56,20,30]
sizes=[400,200,300,100,200,100,300]
plt.scatter(day,no,c=colors,s=sizes,cmap="viridis")    # Color Map
t=plt.colorbar()
t.set_label("Color Bar",fontsize=10)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=10)
plt.ylabel("No",fontsize=10)
plt.show()
```
