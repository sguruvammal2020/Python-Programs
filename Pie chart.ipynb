# Pie Chart

```python
# Creating Pie Chart
import matplotlib.pyplot as plt
x=[10,20,30,40]
plt.pie(x)
plt.show()
```

```python
# Adding names for pie
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
plt.pie(x,labels=y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.4,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex)   # "C" is separated from Pie
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","g","b","y"]
plt.pie(x,labels=y,explode=ex,colors=c)   # Changing default color
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","g","b","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%")   # Adding the symbol % to represent percentage
plt.show()
```

```python
# Adding shadow parameters
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","g","b","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True)   # Adding shadow
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","g","b","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1.5)   
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","g","b","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1.5,labeldistance=1.3)   
plt.show()
```

```python
# Rotate 90 degree in counter clockwise
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","g","b","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1.5,labeldistance=1.3,startangle=90)        
plt.show()
```

```python
# Adding Text property
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","g","b","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1.5,labeldistance=1.3,
        startangle=90,textprops={"fontsize":15})        
plt.show()
```

```python
# Rotate clock wise by counterclock=False
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","g","b","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1.5,labeldistance=1.3,
        startangle=90,textprops={"fontsize":15},counterclock=False)        
plt.show()
```

```python
# Adding line width and color of edge
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","g","b","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1.5,labeldistance=1.3,
        startangle=90,textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":4,"edgecolor":"m"})        
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","g","b","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1.5,labeldistance=1.3,
        startangle=90,textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":4,"edgecolor":"m"},
       center=(3,6))        
plt.show()
```

```python
#Rotate labels
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["C","C++","Java","Python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","g","b","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=0.5,labeldistance=1.9,
        startangle=90,textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":4,"edgecolor":"m"},
       center=(3,6),rotatelabels=False)    
plt.title("Language Pie Chart")
plt.legend(loc=2)
plt.show()
```

```python
# Donut  Pie Chart
import matplotlib.pyplot as plt
x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","c++","Java","Python"]
c=["r","g","b","y"]
plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,radius=1,colors=c)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","c++","Java","Python"]
c=["r","g","b","y"]
plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors="w")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","c++","Java","Python"]
c=["r","g","b","y"]
plt.pie(x,labels=y,radius=1.5)
cr=plt.Circle(xy=(0,0),radius=1,facecolor="w")
plt.gca().add_artist(cr)
plt.show()
```
